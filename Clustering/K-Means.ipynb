# K-Means Clustering

K-means is an unsupervised algorithm used to group data points into clusters based on similarity. 

Imagine you have a bunch of points (or students with different scores) and you want to group them based on how similar they are. K-means helps by finding "centers" (or centroids) that best represent each group.

**How it Works:**

- Initialization: Randomly choose k centers.
- Assignment: For each point, find the closest center and assign the point to that cluster.
- Update: Recalculate each center as the average of all points in its cluster.
- Iteration: Repeat the assignment and update steps until the centers stabilize (they stop moving significantly).

**Goal:** Minimize the sum of the squared distances between points and their cluster centers, which means the points in each cluster are as close to their center as possible.

## Pipeline for applying K-Means

**Data Preprocessing:**
- Load the Data: Import your dataset (e.g., from a CSV file).
- Feature Selection: Choose the features (columns) you want to cluster.
- Scaling: Standardize or normalize the data if features have different units.

**Choosing the Number of Clusters (k):**
- Use the elbow method: Run K-means for a range of k values and plot the within-cluster sum of squares (WCSS). Look for the "elbow" point where adding more clusters doesnâ€™t improve much.

**Fitting the Model:**
- Run the K-means algorithm with the chosen k.
- Get the cluster assignments for each data point.

**Evaluation and Visualization:**
- Visualize the clusters (if possible, reduce dimensions for plotting).
- Evaluate how well the clustering performed, for example by checking the distribution of points in clusters or using metrics like the silhouette score.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('darkgrid')
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans = kmeans.fit(X)
clusters = kmeans.predict(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=clusters, s=50, cmap='Set2')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```

```python
from sklearn.datasets import make_moons
X, y = make_moons(200, noise=.05, random_state=0)
kmeans = KMeans(2, random_state=0).fit(X)
labels = kmeans.predict(X)
centers = kmeans.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)
plt.scatter(X[:, 0], X[:, 1], c=clusters, s=60, cmap='viridis');
plt.show()
```
